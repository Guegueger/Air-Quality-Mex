```python
data_cont = data.query('Parametro == "SO2" or Parametro == "O3" or Parametro == "NO2"')
```

```python
# Crear una nueva columna con solo la fecha (sin horas)
data_cont['Fecha_Dia'] = data_cont['Fecha'].dt.date
data_cont['Fecha_Dia'] = pd.to_datetime(data_cont['Fecha_Dia'])
_ = data_cont.query('Fecha_Dia == "2023-01-01" and Parametro == "NO2"')
_.head(48)  
```

```python
##### Obtenemos el maximo de cada contaminante por dia
data_cont_max = data_cont.groupby(['Fecha_Dia', 'Parametro', 'Estación','Latitud', 'Longitud'])['Concentraciones horarias'].max().reset_index()
data_cont_max.rename(columns={'Concentraciones horarias':'Concentración por día', 'Fecha_Dia': 'Fecha'}, inplace=True)
data_cont_max.head(10)
```

### Definamos una función para categorizar los datos de acuerdo al indice de calidad del aire

```python
def clasificar_calidad_aire(df, columna_concentracion, bueno :None , aceptable:tuple, malo :tuple, muy_malo: tuple, extremadamente_malo:None):
    """
    Clasifica la calidad del aire según la concentración por día y asigna id, calidad del aire y nivel de riesgo.

    Parámetros:
        df (pd.DataFrame): DataFrame con la columna de concentración diaria.
        columna_concentracion (str): Nombre de la columna con la concentración por día.
        Umbrales de concentración para cada nivel de calidad del aire y riesgo
    Retorna:
        pd.DataFrame: DataFrame con columnas adicionales de id, calidad del aire y riesgo.
    """
    # Definir condiciones y valores correspondientes
    condiciones = [
        (df[columna_concentracion] <= bueno),
        (df[columna_concentracion] > aceptable[0]) & (df[columna_concentracion] < aceptable[1]),
        (df[columna_concentracion] >= malo[0]) & (df[columna_concentracion] < malo[1]),
        (df[columna_concentracion] >= muy_malo[0]) & (df[columna_concentracion] < muy_malo[1]),
        (df[columna_concentracion] >= extremadamente_malo)
    ]

    ids = [0, 1, 2, 3, 4]
    calidad_aire = ['Buena', 'Aceptable', 'Mala', 'Muy mala', 'Extremadamente mala']
    riesgos = ['Bajo', 'Moderado', 'Alto', 'Muy alto', 'Extremadamente alto']

    # Asignar valores a nuevas columnas usando np.select
    df['id'] = np.select(condiciones, ids, default=-1)
    df['Calidad del aire'] = np.select(condiciones, calidad_aire, default='Desconocida')
    df['Riesgo'] = np.select(condiciones, riesgos, default='Desconocido')

    return df

```

```python

data_threshold = pd.DataFrame()
contaminantes = ['SO2', 'O3', 'NO2']
for contaminante in contaminantes:
    df = data_cont_max.query('Parametro == @contaminante')
    if contaminante == 'SO2':
        df = clasificar_calidad_aire(df, 'Concentración por día', 0.035, (0.035, 0.075), (0.075, 0.185), (0.185, 0.304), 0.304)
    elif contaminante == 'O3':
        df = clasificar_calidad_aire(df, 'Concentración por día', 0.058, (0.058, 0.090), (0.090, 0.135), (0.135, 0.175), 0.175)
    elif contaminante == 'NO2':
        df = clasificar_calidad_aire(df, 'Concentración por día', 0.053, (0.053, 0.106), (0.106, 0.160), (0.160, 0.213), 0.213)
    data_threshold = pd.concat([data_threshold, df], ignore_index=True, axis=0)
```
